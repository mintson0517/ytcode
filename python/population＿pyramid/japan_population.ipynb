```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Excelファイルのパス
xlsx_file_path = 'japan_population.xlsx'  # 実際のファイルパスを指定

# Excelファイルを読み込む（適切なシート名またはインデックスを指定）
df = pd.read_excel(xlsx_file_path, sheet_name=0)

# 列名を文字列に変更（もし数値であれば）

df.columns = df.columns.astype(str)

# データの確認（欠損値がないかなど）
print(df.head())

```

```python
# グラフのサイズを指定
plt.figure(figsize=(12, 8))

# 1935年の人口ピラミッド
plt.barh(df['年齢'], -df['1935_男'], label='1935年 男性', alpha=0.7, color='#1f77b4')
plt.barh(df['年齢'], df['1935_女'], label='1935年 女性', alpha=0.7, color='#ff7f0e')

# ラベルとタイトル
plt.xlabel('人口')
plt.ylabel('年齢')
plt.title('1935年の人口ピラミッド')
plt.legend()
plt.grid(axis='x', linestyle='-')

# X軸のラベルを調整（男性を左側、女性を右側に）
max_population = int(max(df['1935_男'].max(), df['1935_女'].max()))  # 整数型に変換
tick_values = range(-max_population, max_population + 1, 1000000)  # 100万人単位で設定
plt.xticks(tick_values, [f"{abs(x)//10000}万" for x in tick_values])  # 数値を「万」単位で表示

# グラフの表示
plt.show()
```
