```python
# height_histogram.py

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
mean_height = 158.066166666667  # Mean Height
std_dev = 5.13328579085299     # Standard Deviation
sample_size = 600               # Sample Siz)
```

```python
# Generates data that follows a normal distribution
heights = np.random.normal(loc=mean_height, scale=std_dev, size=sample_size)
```

```python
# Draw a histogram
plt.figure(figsize=(9, 5))
plt.hist(heights, bins=30, color='skyblue', edgecolor='black')

# Title and labels
plt.title('Histogram of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Sample Frequency')

# Grid for better readability
plt.grid()

# Show plot
plt.show()

# 動画解説リンク:
# このコードの解説は以下の動画で確認できます：
# https://youtu.be/Jp1_JlYjH7k
# 動画では、データの生成からヒストグラムの描画までの過程を詳細に説明しています。
```
